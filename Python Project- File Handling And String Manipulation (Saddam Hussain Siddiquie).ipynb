# Python project(part 1)

```python
#Opening text files.
text_open = open('C:\\Users\\sadda\\Downloads\\Python\\Python Project\\conv.txt')
```

```python
#Using readlines in text files.
text_rl = text_open.readlines()
text_rl
```

```python
list_actors = []
#Ignoring '\n' and creating list by splitting with delimiter ':'.
for line in text_rl:
    if line == '\n':
        pass
    else :
        list_actors.append(line.split(':')[0])
#Creating set to avoid repeating.
print('List of unique actors:', set(list_actors))
print('NO : of actors:', len(set(list_actors)))
```

# Python project(part2)

```python
#Creating list of both actor and conversation with delimiter ':'.
list_both = []
for line in text_rl:
    if line == '\n':
        pass
    else :
        list_both.append(line.split(':'))
print(list_both)        
```

```python
#Creating conversation string by splitting with delimiter ':'.
actors_set = set(list_actors)       
for actor in actors_set:
    conv_str = ""
    for line in list_both:
        if line[0] == actor:
            conv_str = conv_str + str(line[1])
#Replacing "!", ".", "’s", ",", "?" with "" and lowercase alphabets of conv_str.      
    conv_str = conv_str.replace("!", "").replace(".", "").replace("’s", "").replace(",", "").replace("?", "").lower()
    conv_split = conv_str.split()
#Creating conversation set.    
    conv_set = set(conv_split)
#Opening file in 'w' mode and creating file of unique words by each actors.
    file = open(actor+'.txt','w')
    for i in conv_set:
        file.write(i + '\n')
    file.close()
text_open.close()      
```
